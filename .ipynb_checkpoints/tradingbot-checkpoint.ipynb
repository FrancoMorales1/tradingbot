```python
import pandas as pd
import numpy as np
df = pd.read_csv('data.csv')
col = ['Timestamp','Open','High','Low','Close']
n = int((df.shape[0]-100000))
d = []
result = []
j=0
for i in range((df.shape[0]-50000),df.shape[0]-4):
    d.append(np.asarray(df.iloc[i:i+3,1]))
    d[j][np.isnan(d[j])]=0

    maxValue = max(d[j])
    if (maxValue < df.iloc[i+4,4]):
        result.append('Comprar')
    elif (maxValue == df.iloc[i+4,4]):
        result.append('Nada')
    else:
        result.append('Vender')
    j=j+1
print(d)
```

```python
X = d
y = result
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
